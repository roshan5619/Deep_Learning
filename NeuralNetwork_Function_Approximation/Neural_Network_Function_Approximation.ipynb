```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tensorflow as tf
from sklearn.preprocessing import StandardScaler

# Load data from part A (assume these functions are already defined)
# If running separately, you'll need to include the create_*_dataset functions here
x1_train, x2_train, y_train = create_training_dataset()
x1_test1, x2_test1, y_test1 = create_test1_dataset()
x1_test2, x2_test2, y_test2 = create_test2_dataset()

# Normalize the data as suggested in the hint
def normalize_data():
    # Create scalers
    scaler_x1 = StandardScaler()
    scaler_x2 = StandardScaler()
    scaler_y = StandardScaler()

    # Reshape data for sklearn scalers
    x1_train_reshaped = x1_train.reshape(-1, 1)
    x2_train_reshaped = x2_train.reshape(-1, 1)
    y_train_reshaped = y_train.reshape(-1, 1)

    # Fit scalers on training data
    scaler_x1.fit(x1_train_reshaped)
    scaler_x2.fit(x2_train_reshaped)
    scaler_y.fit(y_train_reshaped)

    # Transform training data
    x1_train_norm = scaler_x1.transform(x1_train_reshaped).flatten()
    x2_train_norm = scaler_x2.transform(x2_train_reshaped).flatten()
    y_train_norm = scaler_y.transform(y_train_reshaped).flatten()

    # Transform test data
    x1_test1_norm = scaler_x1.transform(x1_test1.reshape(-1, 1)).flatten()
    x2_test1_norm = scaler_x2.transform(x2_test1.reshape(-1, 1)).flatten()
    y_test1_norm = scaler_y.transform(y_test1.reshape(-1, 1)).flatten()

    x1_test2_norm = scaler_x1.transform(x1_test2.reshape(-1, 1)).flatten()
    x2_test2_norm = scaler_x2.transform(x2_test2.reshape(-1, 1)).flatten()
    y_test2_norm = scaler_y.transform(y_test2.reshape(-1, 1)).flatten()

    return (x1_train_norm, x2_train_norm, y_train_norm,
            x1_test1_norm, x2_test1_norm, y_test1_norm,
            x1_test2_norm, x2_test2_norm, y_test2_norm,
            scaler_x1, scaler_x2, scaler_y)

# Get normalized data
(x1_train_norm, x2_train_norm, y_train_norm,
 x1_test1_norm, x2_test1_norm, y_test1_norm,
 x1_test2_norm, x2_test2_norm, y_test2_norm,
 scaler_x1, scaler_x2, scaler_y) = normalize_data()

# Reset default graph
tf.compat.v1.reset_default_graph()

# Disable eager execution
tf.compat.v1.disable_eager_execution()
# B.1) Neural Network with Sigmoid activation
# Create placeholders
x1 = tf.compat.v1.placeholder(tf.float32, shape=[None], name="x1")
x2 = tf.compat.v1.placeholder(tf.float32, shape=[None], name="x2")
y_true = tf.compat.v1.placeholder(tf.float32, shape=[None], name="y_true")

# Combine inputs
inputs = tf.stack([x1, x2], axis=1)

# Create neural network architecture
# First hidden layer with sigmoid activation
weights1 = tf.Variable(tf.compat.v1.truncated_normal([2, 16], stddev=0.1), name="weights1")
biases1 = tf.Variable(tf.zeros([16]), name="biases1")
hidden1 = tf.matmul(inputs, weights1) + biases1
hidden1_activated = tf.nn.sigmoid(hidden1)  # Sigmoid activation

# Second hidden layer with sigmoid activation
weights2 = tf.Variable(tf.compat.v1.truncated_normal([16, 16], stddev=0.1), name="weights2")
biases2 = tf.Variable(tf.zeros([16]), name="biases2")
hidden2 = tf.matmul(hidden1_activated, weights2) + biases2
hidden2_activated = tf.nn.sigmoid(hidden2)  # Sigmoid activation

# Output layer with linear activation
weights3 = tf.Variable(tf.compat.v1.truncated_normal([16, 1], stddev=0.1), name="weights3")
biases3 = tf.Variable(tf.zeros([1]), name="biases3")
y_pred = tf.matmul(hidden2_activated, weights3) + biases3
y_pred = tf.squeeze(y_pred)

# Define loss function (Mean Squared Error)
loss = tf.reduce_mean(tf.square(y_pred - y_true))

# Define optimizer
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=0.001)
train_op = optimizer.minimize(loss)

# Create session and initialize variables
sess = tf.compat.v1.Session()
sess.run(tf.compat.v1.global_variables_initializer())

# Train the network
epochs = 5000
losses = []

for epoch in range(epochs):
    # Run one training step
    _, loss_val = sess.run(
        [train_op, loss],
        feed_dict={
            x1: x1_train_norm,
            x2: x2_train_norm,
            y_true: y_train_norm
        }
    )

    losses.append(loss_val)

    # Print progress
    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss_val:.6f}")

# Plot training progress
plt.figure(figsize=(10, 6))
plt.plot(losses)
plt.title('Training Loss (Sigmoid activation)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.savefig('sigmoid_training_loss.png')
# plt.close()
plt.show()
# Calculate predictions
# Training dataset
y_train_pred_norm = sess.run(y_pred, feed_dict={x1: x1_train_norm, x2: x2_train_norm})
y_train_pred = scaler_y.inverse_transform(y_train_pred_norm.reshape(-1, 1)).flatten()
train_mse = np.mean((y_train - y_train_pred) ** 2)

# Test1 dataset
y_test1_pred_norm = sess.run(y_pred, feed_dict={x1: x1_test1_norm, x2: x2_test1_norm})
y_test1_pred = scaler_y.inverse_transform(y_test1_pred_norm.reshape(-1, 1)).flatten()
test1_mse = np.mean((y_test1 - y_test1_pred) ** 2)

# Test2 dataset
y_test2_pred_norm = sess.run(y_pred, feed_dict={x1: x1_test2_norm, x2: x2_test2_norm})
y_test2_pred = scaler_y.inverse_transform(y_test2_pred_norm.reshape(-1, 1)).flatten()
test2_mse = np.mean((y_test2 - y_test2_pred) ** 2)

# Print results
print("\nResults with Sigmoid activation:")
print(f"Training MSE: {train_mse:.2f}")
print(f"Test1 MSE: {test1_mse:.2f}")
print(f"Test2 MSE: {test2_mse:.2f}")

# Plot predictions for Training dataset
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x1_train, x2_train, y_train_pred, c=y_train_pred, cmap='viridis', s=15)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('Neural Network Predictions - Training (Sigmoid)')
fig.colorbar(scatter, ax=ax, label='Predicted Y')
plt.tight_layout()
plt.savefig('sigmoid_train_predictions.png')
# plt.close()
plt.show()

# Plot predictions for Test1 dataset (subset for clarity)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sample_indices = np.arange(0, len(x1_test1), 10)
scatter = ax.scatter(x1_test1[sample_indices], x2_test1[sample_indices],
                    y_test1_pred[sample_indices], c=y_test1_pred[sample_indices], cmap='viridis', s=15)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('Neural Network Predictions - Test1 (Sigmoid)')
fig.colorbar(scatter, ax=ax, label='Predicted Y')
plt.tight_layout()
plt.savefig('sigmoid_test1_predictions.png')
# plt.close()
plt.show()

# Plot predictions for Test2 dataset (subset for clarity)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sample_indices = np.arange(0, len(x1_test2), 10)
scatter = ax.scatter(x1_test2[sample_indices], x2_test2[sample_indices],
                    y_test2_pred[sample_indices], c=y_test2_pred[sample_indices], cmap='viridis', s=15)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.set_title('Neural Network Predictions - Test2 (Sigmoid)')
fig.colorbar(scatter, ax=ax, label='Predicted Y')
plt.tight_layout()
plt.savefig('sigmoid_test2_predictions.png')
# plt.close()
plt.show()
# # Close the session
sess.close()
```
